# 01 - Data Exploration

Starter notebook for legal, leakage-safe international football modeling.

Checks:
- source columns and nulls
- chronological ordering
- outcome distribution

```python
import pandas as pd

matches = pd.read_csv('../data/processed/matches_clean.csv')
features = pd.read_csv('../data/processed/features.csv')

print('Matches shape:', matches.shape)
print('Features shape:', features.shape)
print(features[['date', 'home_team', 'away_team', 'target']].head())
```

```python
features['date'] = pd.to_datetime(features['date'])
features = features.sort_values('date')

print('Date range:', features['date'].min(), 'to', features['date'].max())
print('\nOutcome distribution:')
print(features['target'].value_counts(normalize=True).rename('share'))
```
